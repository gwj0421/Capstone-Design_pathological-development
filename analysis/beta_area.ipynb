```python
# OpenCV 패키지 사용
import cv2 as cv
import matplotlib.pyplot as plt  # 시각화 도구
import numpy as np
import matplotlib.pyplot as plt
import os
import glob
```

```python
filename= '0552'
workspace='D:/Codes/testing/Development-of-histologic-and-pathological-photographic-intellectualization-analysis-system/test'
dataset=os.path.join(workspace,'dataset')
img_list=glob.glob(os.path.join(dataset,'Image_labeling/'+filename+'_labeling')+'/*.jpg')
```

```python
len(img_list)
```

```python
img_color = cv.imread(img_list[1476])
img_gray = cv.cvtColor(img_color, cv.COLOR_BGR2GRAY)
ret, img_binary = cv.threshold(img_gray, 127, 255, 0)
contours, hierarchy = cv.findContours(img_binary,  cv.RETR_EXTERNAL, cv.CHAIN_APPROX_SIMPLE)
plt.figure(figsize=(15, 15))
plt.subplot(131)
plt.imshow(img_color)
# for cnt in contours:
#     cv.drawContours(img_color, [cnt], 0, (255, 255, 255), -1)  # blue
cv.drawContours(img_color, contours[1], 0, (0, 255, 255), 3)  # blue
plt.subplot(132)
plt.imshow(img_color)

area=0
for cnt in contours:
    area += cv.contourArea(cnt)

print(area)
plt.subplot(133)
plt.imshow(img_color)
```
