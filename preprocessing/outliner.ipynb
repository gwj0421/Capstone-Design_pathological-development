```python
# OpenCV 패키지 사용
import cv2
import matplotlib.pyplot as plt  # 시각화 도구
import numpy as np
from PIL import Image
import tqdm
import matplotlib.pyplot as plt
import os
```

```python
workspace='D:/Codes/testing/Development-of-histologic-and-pathological-photographic-intellectualization-analysis-system/test'
images=os.listdir(os.path.join(workspace,'dataset/source'))
images=[workspace+'/dataset/source/'+DATA for DATA in images]
```

```python
images[9]
```

```python
original=cv2.imread(images[9])
plt.imshow(original)
```

```python
temp=np.copy(original)
imgBlur = cv2.medianBlur(temp, 55)   
imgBlur2 = cv2.cvtColor(imgBlur, cv2.COLOR_BGR2RGB)
# plt.figure(figsize=(20,21))
# plt.imshow(imgBlur2)
```

```python
flag=np.array((imgBlur2[:,:,0]>200)&(imgBlur2[:,:,1]>200)&(imgBlur2[:,:,2]>200))
flag.shape
```

```python
temp=np.copy(imgBlur2)
temp[flag,:]=[0,0,0]
plt.figure(figsize=(20,21))
plt.imshow(temp)
```

```python
temp.shape
```

```python
img_gray=np.copy(temp)
img_gray=cv2.cvtColor(img_gray,cv2.COLOR_BGR2GRAY)
contours,_ = cv2.findContours(img_gray, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
```

```python
cn=np.copy(temp)
cn=cv2.drawContours(cn, contours, -1, (0,0,255), 5)
plt.figure(figsize=(20,21))
plt.imshow(cn)
```

```python
cn_filled=np.copy(temp)
cn_filled=cv2.drawContours(cn_filled, contours, -1, (255,255,0), -1)
plt.figure(figsize=(20,21))
plt.imshow(cn_filled)
```

```python
area=0
for cnt in contours:
    area += cv2.contourArea(cnt)

print(area)
```
